### Look at the Big Picture
The dataset comes from the UCI Machine Learning Repository and measures eleven physicochemical parameters including pH and alcohol content of 1,599 different red wines. The quality of each wine has been judged by human judges. And the scores range from zero to ten with zero for the worst quality and ten for the best quality. The data contain the following explanatory variables: fixed acidity, volatility acidity, citric acid, residual sugar, chlorides, free sulfur dioxide, total sulfur dioxide, density, pH, sulphates and alcohol content. Understanding these attributes could provide us with insight into model design, and domain expertise is often important for designing successful machine learning systems. But for this example, it will not be necessary to interpret the effects of physicochemical attributes of wine quality. For the sake of brevity, therefore, we will delete some explanatory variables.

### Visualize the data to gain insights

Visualizing the data can help indicate whether relationships exist between the
response variable and the explanatory variables. Let's use to create some scatter plots.

```python
plt.scatter(dataset['alcohol'], dataset['quality'])
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.title('Alcohol VS Quality')
plt.show()
```

A weak positive relationship between Alcohol content and Quality is visible in the scatter plot in the previous figure; wines that have high alcohol content are often of high quality.

```python
plt.scatter(dataset['volatile acidity'], dataset['quality'])
plt.xlabel('Volatile Acidity')
plt.ylabel('Quality')
plt.title('Volatile Acidity VS Quality')
plt.show()
```

The previous figure reveals a negative relationship between Volatile Acidity and Quality.

These plots suggest that the response variable depends on multiple explanatory variables; let's model the relationship with multiple linear regression. How can we decide which explanatory variables to include in the model?

### Dataframe.corr()

The correlation matrix confirms that the strongest positive correlation is

### Preparation the data for Machine Learning algoritms

####  Create variables : x and y

```python
x = dataset[list(dataset.columns)[:-1]]
y = dataset['quality']
```

#### Visualize x and y

```python
x
```

```python
y
```

### Divide the dataset in Training / Test Set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

### Fitting and evaluating the model

Now we will split the data into training and testing sets, train regressor, and evaluate its predictions :

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_predictions = regressor.predict(x_test)
print('R-squared: %s'  %  regressor.score(x_test, y_test))
```

### Cross Validation

We can use cross validation to produce a better estimate of the estimator's performance.

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(regressor, x, y, cv=5)
```

```python
print(scores.mean())
print(scores)
```

The  helper function allows us to easily perform cross validation using the provided data and estimator. We specified five-fold cross validation using the keyword argument. That is, each instance will be randomly assigned to one of five
partitions. Each partition will be used to train and test the model. The function returns the value of the estimator's score method for each round.
